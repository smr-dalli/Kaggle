```
models = {'RandomForestClassifier':RandomForestClassifier(),'RandomForestRegressor':RandomForestRegressor(),'LinearRegression':LinearRegression(),'LogisticRegression':LogisticRegression(),'DecisionTreeClassifier':DecisionTreeClassifier(),'DecisionTreeRegressor':DecisionTreeRegressor(),
'XGBClassifier':XGBClassifier(use_label_encoder=False),'XGBRegressor':XGBRegressor(),'AdaBoostClassifier':AdaBoostClassifier(),'AdaBoostRegressor':AdaBoostRegressor(),
'CatBoostClassifier':CatBoostClassifier(verbose=False),'CatBoostRegressor':CatBoostRegressor(verbose=False),'LGBMClassifier':LGBMClassifier()}
### split the data
X_train,X_test,y_train,y_test = train_test_split(num_cols,y,test_size=0.2,stratify=y,random_state=42)
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```
#pipe = make_pipeline(preprocesing,model)
for model_name,model in models.items():
    #model_name = [RandomForestClassifier,RandomForestRegressor,LinearRegression,LogisticRegression,DecisionTreeClassifier,DecisionTreeRegressor]
    model.fit(X_train,y_train)
    ypred = model.predict(X_test)
    error1 = mean_absolute_error(ypred,y_test)
    error2 = mean_squared_error(ypred,y_test)
    error3 = np.sqrt(mean_squared_error(ypred,y_test))
    print(f'The mean absolute error of {model_name} is : {error1:.2f} ')
    print(f'The mean_squared_error of {model_name} is : {error2:.2f} ')
    print(f'The root mean_squared_error of {model_name} is : {error3:.2f} ')
    print()
```

```
num_preprocessing = Pipeline(steps=[('scaler',StandardScaler())])
cat_preprocessing = Pipeline(steps=[('ohe',OneHotEncoder())])

preprocesing = ColumnTransformer(('num',num_preprocessing,num_cols_list),
('cat',cat_preprocessing,cat_cols_list),remainder='passthrough')
preprocesing
```

```
rand_mdl = RandomForestClassifier()
pipe = Pipeline(steps = [('pre',preprocesing),('mod',rand_mdl)])
pipe
```
